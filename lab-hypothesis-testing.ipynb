```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df.tail(60)
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
df_dragon = df[df["Type 1"]=="Dragon"]["HP"]
df_grass = df[df["Type 1"]=="Grass"]["HP"]
```

```python
alpha = 0.05
stat, p_value = st.ttest_ind(df_dragon,df_grass, equal_var=False)
#H0= Mu df1 = Mu df2 , H1= Mu df1 =! Mu df2
if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import math
```

```python
def distancia_euclidiana(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

school_coords = (-118, 37)
hospital_coords = (-122, 34)

df['distancia_escuela'] = df.apply(lambda row: distancia_euclidiana(row['longitude'], row['latitude'], school_coords[0], school_coords[1]), axis=1) 
df['distancia_hospital'] = df.apply(lambda row: distancia_euclidiana(row['longitude'], row['latitude'], hospital_coords[0], hospital_coords[1]), axis=1)
   
```

```python
df_cerca_escuela = df[df["distancia_escuela"]<0.5]["median_house_value"]
df_lejos_escuela = df[df["distancia_escuela"]>0.5]["median_house_value"]
```

```python
alpha=0.05
```

```python
stat, p_value = st.ttest_ind(df_cerca_escuela,df_lejos_escuela,equal_var=False, alternative="less")
#H0= Mu df1 >= Mu df2 , 
# H1= Mu df1 < Mu df2
if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```
